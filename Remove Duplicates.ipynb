```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv(r"C:\Users\shant\Desktop\SIH\MCQ Scraping\sanfoundry-scraper-master\html to csv/400_balanced_with_difficulty.csv")
```

```python
corpus = list(df["Question"].values)

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
threshold = 0.9

for x in range(0,X.shape[0]):
  for y in range(x,X.shape[0]):
    if(x!=y):
      if(cosine_similarity(X[x],X[y])>threshold):
        print(corpus[x])
        print(corpus[y])
        print("Cosine similarity:",cosine_similarity(X[x],X[y]))
        print()
```
